```python
#importing libraries
import numpy as np
import math
import statistics as st
```

```python
# creating data set
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]
data_arr=np.array(data) # converting dataset into a numpy array
X=data_arr[:,:3].astype(float) # Converting the type of features from sting to float
y_arr=data_arr[:,3] # True labels of the data

# One hot encoding
for i in range(len(y_arr)):
  if y_arr[i]=='Apple':
    y_arr[i]='010'
  elif y_arr[i]=='Banana':
    y_arr[i]='100'
  else:
    y_arr[i]='001'
y=y_arr
```

```python
# Euclidean distance
def euclidian_dis(x1,x2):
  return np.sqrt(np.sum((np.array(x1)-np.array(x2))**2))

# Manhattan distance
def manhattan(x1,x2):
  return np.sum(np.abs(np.array(x1)-np.array(x2)))

# Minkowski distance
def Minkowski(x1,x2,p):
  return np.power(np.sum((np.array(x1)-np.array(x2))**p),1/p)

# KNN Class
class KNN:
  def __init__(self,k=3):
    self.k=k
  def fit(self,X,y): # storing the training dataset
    self.X=X
    self.y=y
  def predict(self,X_test): # predicting the categories
    pre=[]
    for i in range(X_test.shape[0]):
      lis=[]
      for j in range((self.X).shape[0]):
        # lis.append(euclidian_dis(self.X[j],X_test[i])) # Euclidian distance

        # lis.append(manhattan(self.X[j],X_test[i])) # manhattan distance

        lis.append(Minkowski(self.X[j],X_test[i],4)) # minkowski at p=4

      lis2=lis.copy()
      lis2.sort()
      lis1=[]
      for a in lis2[:self.k]:
        for j in range(len(lis)):
          if a==lis[j]:
            lis1.append(self.y[j])
            break
      pre.append(st.mode(lis1))
    return pre
  def predict_one(self,x): # predicting category of only one point
    lis=[]
    for j in range((self.X).shape[0]):
      lis.append(euclidian_dis(self.X[j],x))
    lis2=lis.copy()
    lis2.sort()
    lis1=[]
    for a in lis2[:self.k]:
      for j in range(len(lis)):
        if a==lis[j]:
          lis1.append(self.y[j])
          break
    return st.mode(lis1)


```

```python
model=KNN(k=3) # Creating instance of the data
test_data = np.array([
    [118, 6.2, 0],
    [160, 7.3, 1],
    [185, 7.7, 2]
])
model.fit(X,y) # storing the data
predictions=model.predict(test_data) # making predictions
print(predictions)
# 100 => Banana
# 010 => Apple
# 001 => Orange
```

```python
y_true=['100','010','001']
accuracy=0
for i in range(len(y_true)): # Checking accuracy of my predictions
  accuracy+=(y_true[i]==predictions[i])
accuracy=(accuracy/len(y_true))
print(accuracy)
```
